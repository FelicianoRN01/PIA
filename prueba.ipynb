```python
import mysql.connector

mydb = mysql.connector.connect(
host="localhost",
user="root",
password="1234",
database="personas"
)

mycursor = mydb.cursor()

mycursor.execute("SELECT * FROM personas")

myresult = mycursor.fetchall()

for x in myresult:
    print(x)
```
