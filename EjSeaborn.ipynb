```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

# Dado el dataset con la información de los pasajeros del titanic que puedes cargar con el código anterior
# realiza los siguientes ejercicios (realizando gestión de nulos):
```

1. Representa mediante un histograma (utilizando sns.distplot) la tarifa pagada por los pasajeros (fare). Utiliza 20 bins y muestra dos gráficas, una con la estimación de la probabilidad y otra sin ella.
- Para crear dos ejes dentro de un gráfico deberás hacer uso de matplotlib.


```python
# Gestión de nulos en la columna 'fare'
titanic['fare'].fillna(titanic['fare'].median(), inplace=True)

# Crear los subgráficos
fig, axes = plt.subplots(1, 2, figsize=(15, 6), sharey=True)

# Histograma con la estimación de probabilidad (kde=True)
sns.histplot(titanic['fare'], bins=20, kde=True, ax=axes[0])
axes[0].set_title("Histograma con estimación de probabilidad")
axes[0].set_xlabel("Tarifa (fare)")
axes[0].set_ylabel("Frecuencia")

# Histograma sin la estimación de probabilidad (kde=False)
sns.histplot(titanic['fare'], bins=20, kde=False, ax=axes[1])
axes[1].set_title("Histograma sin estimación de probabilidad")
axes[1].set_xlabel("Tarifa (fare)")

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

2. Representa la distribución de edades de los pasajeros con un histograma (utilizando sns.kdeplot). Incluye el parámetro hue para diferenciar por la variable sex.


```python
# Gestión de nulos en la columna 'age'
titanic['age'].fillna(titanic['age'].median(), inplace=True)

# Crear el histograma con sns.kdeplot
plt.figure(figsize=(10, 6))
sns.kdeplot(data=titanic, x="age", hue="sex", fill=True, common_norm=False, alpha=0.5)
plt.title("Distribución de edades por sexo")
plt.xlabel("Edad")
plt.ylabel("Densidad")
plt.legend(title="Sexo")
plt.show()
```

3. Representa la relación entre tarifa y edad.
- Usa un diagrama de dispersión con sns.scatterplot.
- Incluye el parámetro hue para diferenciar por survived y size para escalar los puntos según la cantidad de hermanos/esposos (sibsp).

```python
# Gestión de nulos en las columnas 'age' y 'fare'
titanic['age'].fillna(titanic['age'].median(), inplace=True)
titanic['fare'].fillna(titanic['fare'].median(), inplace=True)

# Crear el diagrama de dispersión
plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=titanic, 
    x="age", 
    y="fare", 
    hue="survived", 
    size="sibsp", 
    palette="coolwarm", 
    sizes=(20, 200), 
    alpha=0.7
)

# Etiquetas y título
plt.title("Relación entre tarifa y edad\n(Diferenciando por sobrevivencia y número de hermanos/esposos)")
plt.xlabel("Edad")
plt.ylabel("Tarifa")
plt.legend(title="Leyenda", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

4. Representa la distribución de todas las combinaciones binarias de columnas con sns.pairplot.

```python
# Gestión de nulos en columnas numéricas y categóricas
titanic['age'].fillna(titanic['age'].median(), inplace=True)
titanic['fare'].fillna(titanic['fare'].median(), inplace=True)
titanic['embarked'].fillna(titanic['embarked'].mode()[0], inplace=True)

# Seleccionar las columnas para el pairplot
columns_to_plot = ['age', 'fare', 'survived', 'sibsp', 'parch']

# Crear el pairplot
sns.pairplot(
    data=titanic[columns_to_plot],
    hue='survived',  # Diferenciar por la columna 'survived'
    palette='coolwarm',  # Colores para sobrevivientes y no sobrevivientes
    diag_kind='kde',  # Usar Kernel Density Estimation en la diagonal
    markers=['o', 's']  # Diferentes marcadores para cada categoría de survived
)
```

5. Representa las correlaciones del dataset con un mapa de calor usando sns.heatmap.
- Ten en cuenta que la correlación sólo puede trabajar con valores numéricos.


```python
# Gestión de nulos en columnas numéricas
titanic['age'].fillna(titanic['age'].median(), inplace=True)
titanic['fare'].fillna(titanic['fare'].median(), inplace=True)

# Seleccionar solo las columnas numéricas para calcular la correlación
numerical_columns = titanic.select_dtypes(include='number')

# Calcular la matriz de correlación
correlation_matrix = numerical_columns.corr()

# Crear el mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(
    correlation_matrix, 
    annot=True,  # Mostrar los valores numéricos
    fmt=".2f",   # Formato de los valores numéricos
    cmap="coolwarm",  # Colores del mapa
    linewidths=0.5  # Espacio entre las celdas
)
plt.title("Mapa de calor de correlaciones")
plt.show()
```

6. Representa la relación entre relación entre tarifa (fare), edad (age), clase (pclass) y tipo de persona (who).
- Usa sns.relplot con las variables seleccionadas.
- Usa el parámetro hue para diferenciar por survived.
- Usa los facets para diferenciar por tipo de persona (who).


```python
# Crear el gráfico relplot
sns.relplot(
    data=titanic,
    x="age",
    y="fare",
    hue="survived",
    col="who",
    kind="scatter",
    palette="coolwarm",
    height=5,
    aspect=1
)

# Etiquetas y título
plt.subplots_adjust(top=0.85)
plt.suptitle("Relación entre tarifa, edad, clase y tipo de persona\n(Diferenciando por supervivencia)")
plt.show()
```

7. Representa las relación entre tarifa y edad con un sns.jointplot.
- Usa el tipo kde para visualizar las densidades conjuntas y marginales.


```python
# Crear el gráfico jointplot con tipo kde
sns.jointplot(
    data=titanic,
    x="age",
    y="fare",
    kind="kde",
    fill=True,
    cmap="coolwarm"
)

# Etiquetas y título
plt.suptitle("Relación entre tarifa y edad", y=1.02)
plt.show()
```

8. Representa la edad en función de la clase (pclass)  usando sns.catplot.
- Usa hue para diferenciar por sex.
- Usa los facets para diferenciar por puerto de embarque (embark_town) y si sobrevivio (suvirved).

```python
# Crear el gráfico catplot
sns.catplot(
    data=titanic,
    x="pclass",
    y="age",
    hue="sex",
    col="embark_town",
    row="survived",
    kind="box",
    height=4,
    aspect=1
)

# Etiquetas y título
plt.subplots_adjust(top=0.9)
plt.suptitle("Distribución de edades en función de la clase\n(Diferenciando por sexo, puerto de embarque y supervivencia)")
plt.show()
```

9. Representa el cambio promedio de las tarifas (fare) según la clase utilizando un intervalo de confianza del 90%.
- Usa sns.lineplot para mostrar la relación promedio entre pclass y fare.
- Diferencia por sex con el parámetro hue.


```python
# Crear el gráfico lineplot
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=titanic,
    x="pclass",
    y="fare",
    hue="sex",
    ci=90,
    marker="o"
)

# Etiquetas y título
plt.title("Cambio promedio de las tarifas según la clase\n(Diferenciando por sexo)")
plt.xlabel("Clase (pclass)")
plt.ylabel("Tarifa (fare)")
plt.legend(title="Sexo")
plt.show()
```

10. Utiliza un diagrama de dispersión categórico para comparar la distribución de tarifas (fare) entre clases (pclass).
- Usa sns.stripplot y sns.swarmplot y elige el que mejor transmita la información.
- Usa hue para diferenciar por tipo de persona (who).

```python
# Diagrama de dispersión categórico con sns.stripplot
plt.figure(figsize=(10, 6))
sns.stripplot(
    data=titanic,
    x="pclass",
    y="fare",
    hue="who",
    jitter=True,
    dodge=True,
    palette="coolwarm"
)
plt.title("Distribución de tarifas entre clases (stripplot)")
plt.xlabel("Clase (pclass)")
plt.ylabel("Tarifa (fare)")
plt.legend(title="Tipo de persona")
plt.show()
```

12. Utiliza un diagrama de violín para mostrar las tarifas según clase y sexo.
- Agrega split=True para superponer los datos masculinos y femeninos dentro de cada clase.

```python
# Crear el gráfico de violín
plt.figure(figsize=(10, 6))
sns.violinplot(
    data=titanic,
    x="pclass",
    y="fare",
    hue="sex",
    split=True,
    palette="coolwarm"
)

# Etiquetas y título
plt.title("Distribución de tarifas según clase y sexo")
plt.xlabel("Clase (pclass)")
plt.ylabel("Tarifa (fare)")
plt.legend(title="Sexo")
plt.show()
```

13. Representa el porcentaje de supervivencia por clase y sexo utilizando un diagrama de barras.
- Utiliza los facets para diferenciar por categoría de edad (valores de edad agrupados de 10 en 10). Mostrando como mucho 3 columnas y tantas filas necesites.


```python
# Crear una nueva columna para agrupar las edades en categorías de 10 en 10 años
titanic['age_group'] = pd.cut(titanic['age'], bins=range(0, 101, 10), right=False)

# Crear el gráfico catplot
sns.catplot(
    data=titanic,
    x="pclass",
    y="survived",
    hue="sex",
    col="age_group",
    kind="bar",
    height=4,
    aspect=0.8,
    col_wrap=3
)

# Etiquetas y título
plt.subplots_adjust(top=0.9)
plt.suptitle("Porcentaje de supervivencia por clase y sexo\n(Diferenciando por categoría de edad)")
plt.show()
```

14. Representa una línea de regresión entre la tarifa y la edad.


```python
# Crear el gráfico de regresión
plt.figure(figsize=(10, 6))
sns.regplot(
    data=titanic,
    x="age",
    y="fare",
    scatter_kws={"alpha": 0.5},
    line_kws={"color": "red"}
)

# Etiquetas y título
plt.title("Línea de regresión entre tarifa y edad")
plt.xlabel("Edad")
plt.ylabel("Tarifa")
plt.show()
```
