### 1. Escribe un programa que pida un número al usuario e imprima si es par o impar.


```python
n=(int)(input("Dime un numero:",))
if n%2==0:
    print("Este numero es par")
elif n%1==0:
    print("Este numero es impar")
```

### 2. Escribe un programa que reciba tres números como entrada y determine cuál de ellos es el mayor.

```python
n1=(int)(input("Dime el primer numero: "))
n2=(int)(input("Dime el segundo numero: "))
n3=(int)(input("Dime el tercer numero: "))

if n1>n2>n3:
    print("El primer numero es mayor")
elif n2>n1>n3:
    print("El segundo numero es mayor")
elif n3>n2>n1:
    print("El tercer numero es mayor")
```
