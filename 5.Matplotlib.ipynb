Graficar: $f(x)=x^2−4x+4$

```python
x = np.linspace(-1, 5, 400)
```

```python
y = x**2 - 4*x + 4
```

```python
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='f(x) = x^2 - 4x + 4', color='red')
plt.title('Gráfico de la Función Cuadrática')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

# Uso de subplot

```python
x = np.linspace(0, 10, 100)
y_linear = x
y_exponential = np.exp(x)
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(x, y_linear, 'r-')
plt.title('Función Lineal')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.subplot(1, 2, 2)
plt.plot(x, y_exponential, 'b-')
plt.title('Función Exponencial')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(x, y_linear, label='Linear f(x) = x', color='red', linestyle='-')
plt.plot(x, y_exponential, label='Exponential f(x) = e^x', color='blue', linestyle='--')
plt.ylim(0, 10)
plt.grid(True)
plt.show()
```

# Subplots

```python
# Crear figura y un arreglo de ejes (2x2)
fig, axs = plt.subplots(2, 2, figsize=(10, 10))  # 2 filas, 2 columnas

# Datos
x = np.linspace(0, 2*np.pi, 100)
funcs = [np.sin(x), np.cos(x), np.tan(x), np.sqrt(x)]

# Títulos para cada subgráfico
titles = ['Seno', 'Coseno', 'Tangente', 'Raíz Cuadrada']

# Llenar cada subgráfico con una función
for ax, func, title in zip(axs.ravel(), funcs, titles):
    ax.plot(x, func)
    ax.set_title(title)
    ax.grid(True)

# Ajustar automáticamente el layout de los subgráficos para evitar solapamientos
plt.tight_layout()
plt.show()
```
